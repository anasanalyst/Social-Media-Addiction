```python
#Making some additional features to the group before i start the data analysis
#Creating a Age group so it can be easier while analysis
bins=[15,20,25,30,35]
labels=['16-20','21-25','26-30','31-35']
df['Age_Group']=pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
#Creating Daily usage group
usage_bins = [0, 2, 4, 6, 12, 24]
usage_labels = ['Minimal (0-2h)', 'Moderate (2-4h)', 'High (4-6h)', 'Very High (6-12h)', 'Extreme (12h+)']
df['Usage_Category'] = pd.cut(df['Avg_Daily_Usage_Hours'], bins=usage_bins, labels=usage_labels)
```

```python
# Categorize sleep hours
sleep_bins = [0, 5, 7, 9, 12]
sleep_labels = ['Poor (<5h)', 'Fair (5-7h)', 'Good (7-9h)', 'Excellent (9h+)']
df['Sleep_Category'] = pd.cut(df['Sleep_Hours_Per_Night'], bins=sleep_bins, labels=sleep_labels)
```

```python
#1 Starting the EDA with Age distribution and Gender distribution

#Age Distribution
plt.figure(figsize=(16,6))

plt.subplot(1,2,1)
plt.style.use('fivethirtyeight')
sns.histplot(df,x='Age', kde=True, bins=15)
plt.title("Age Distribution of Students", fontsize=15)
plt.xlabel("Ages", fontsize=12)
plt.ylabel('Counts', fontsize='12')

#Gender Distribution
plt.subplot(1,2,2)
gender_counts=df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution",fontsize=15)
plt.axis('equal')
plt.tight_layout()
```

```python
### Now I'm going to do the country and academic level distribution and for both barplot will be used

plt.figure(figsize=(15,5))
sns.set_context('paper')
#top 10Country Level distribution by average hourly usage
plt.subplot(1,2,1)
top_10countries=df.groupby('Country')['Avg_Daily_Usage_Hours'].max().nlargest(10).reset_index()

sns.barplot(top_10countries, x='Country',y='Avg_Daily_Usage_Hours', palette='Set1')
plt.xticks(rotation=45)
plt.title("TOP COUNTRIES SOCIAL USAGE", fontsize=13)
plt.xlabel("Countries", fontsize=12)
plt.ylabel("AVERAGE HOURS", fontsize=12)


#Academic level distribution
plt.subplot(1,2,2)
academic_dist=df['Academic_Level'].value_counts()
sns.barplot(x=academic_dist.index, y=academic_dist.values,palette='Set2')
plt.tight_layout()
```

```python
## Average Daily Usage Hours
plt.figure(figsize=(12,6))
sns.histplot(df['Avg_Daily_Usage_Hours'], kde=True, bins=20, color='black')
plt.title("Average Daily Social Media Usage")
plt.xlabel("Hours", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.axvline(df['Avg_Daily_Usage_Hours'].mean(), color='red', linestyle='--',
           label=f'Mean: {df["Avg_Daily_Usage_Hours"].mean():.2f} hours')
plt.legend()
plt.show()
```

```python
## Most used platforms
plt.figure(figsize=(15,7))
platform_counts= df['Most_Used_Platform'].value_counts()
sns.barplot(x=platform_counts.index, y=platform_counts.values, palette='magma')
plt.title('Widely Used Platforms')
plt.xlabel("Platforms",fontsize=12)
plt.ylabel("Counts",fontsize=12)
plt.show()
```

```python
## Social media usage by Age and Gender
plt.figure(figsize=(16,7))
plt.style.use('bmh')

#Usage by Age
plt.subplot(1,2,1)
#using the age group I've already made above.
sns.boxplot(df, x='Age_Group', y='Avg_Daily_Usage_Hours', palette='Set1')
plt.title("Social Media Usage by Age")
plt.xlabel("Age_Group")
plt.ylabel('AVG Daily Usage Hours')

#Usage by gender
plt.subplot(1,2,2)

sns.boxplot(df, x='Gender', y='Avg_Daily_Usage_Hours', palette='Set2')
plt.title('Social Media Usage by Gender')
plt.xlabel("Gender")
plt.ylabel('AVG Daily Usage Hours')
plt.show()
```

```python
# 3.Mental Health Score VS Social Media Usage

plt.figure(figsize=(10,6))
sns.scatterplot(df, x='Avg_Daily_Usage_Hours', y='Mental_Health_Score', alpha=0.7, hue='Gender')
plt.title("Mental Health Score VS Social Media Usage", fontsize=15)
plt.xlabel("Daily Average Usage Hours")
plt.ylabel('Mental Health Score')
#ADDING A REGRESSION LINE
sns.regplot(df,x='Avg_Daily_Usage_Hours',y='Mental_Health_Score', scatter=False, color='black')
#Calculating the correlation
correlation=df['Avg_Daily_Usage_Hours'].corr(df['Mental_Health_Score'])
plt.annotate(f'Correlation: {correlation:.2f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=14 )
plt.show()
```

```python
# #### Sleep Hours vs Social Media Usage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avg_Daily_Usage_Hours', y='Sleep_Hours_Per_Night', data=df, alpha=0.6, hue='Gender')
plt.title('Sleep Hours vs Social Media Usage', fontsize=14)
plt.xlabel('Average Daily Usage Hours', fontsize=12)
plt.ylabel('Sleep Hours Per Night', fontsize=12)

# Add regression line
sns.regplot(x='Avg_Daily_Usage_Hours', y='Sleep_Hours_Per_Night', data=df, scatter=False, color='red')

# Calculate correlation
correlation2 = df['Avg_Daily_Usage_Hours'].corr(df['Sleep_Hours_Per_Night'])
plt.annotate(f'Correlation2: {correlation:.2f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12)

plt.show()
```

```python
# Checking out addiction score by sleeping category using boxplot
plt.figure(figsize=(12,5))
sns.boxplot(df,x='Sleep_Category', y='Addicted_Score', palette='Set1' )
plt.title('Sleep Quality by Addiction Score')
plt.xlabel("Sleep Category")
plt.ylabel('Addiction Score')
plt.show()
```

```python
#Relationship status and conflicts Analysis

plt.figure(figsize=(12,5))
sns.boxplot(df, x='Relationship_Status', y='Avg_Daily_Usage_Hours', palette="Set2")
plt.title("Relationship Status by Avg Social Media Usage")
plt.xlabel('Relationship Status')
plt.ylabel('Hourly Usage')
plt.show()
```
